#Problem 1: Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0

    # Calculate total likes using loop
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# Example
likes = [250, 300, 200, 300]
total, status = analyze_engagement(likes)
print("Total Likes:", total)
print("Post Status:", status)
```

# Problem 2: Medicine Stock Alert System

```python
def check_medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    return stock, status


# Example
stock, status = check_medicine_stock(6)
print("Medicine Stock:", stock)
print("Status:", status)
```

#Problem 3: Rainfall Adequacy Checker

```python
def check_rainfall(rainfall_list, required_level):
    total = 0

    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status


# Example
rainfall = [70, 75, 80, 65]
avg, status = check_rainfall(rainfall, 70)
print("Average Rainfall:", avg)
print("Rainfall Status:", status)
```

# Problem 4: Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    return result


# Example
users = ["raj", "anita", "raj", "meena"]
print("Duplicate Accounts Found:", detect_duplicate_accounts(users))
```

# Problem 5: Appointment Eligibility Checker

```python
def check_appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return age, status


# Example
age, status = check_appointment_eligibility(21)
print("Patient Age:", age)
print("Eligibility Status:", status)
```

# Problem 6: Premium Crop Price Filter


```python
def filter_premium_crops(price_list):
    premium = []

    for price in price_list:
        if price > 2000:
            premium.append(price)

    return premium


# Example
prices = [1500, 2500, 3200, 1800]
print("Premium Crops:", filter_premium_crops(prices))
```

# Problem 7: Application Health Checker

```python
def check_application_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return errors, status


# Example
errors, status = check_application_health(7)
print("Error Count:", errors)
print("System Status:", status)
```

#Problem 8: Daily Transaction Limit Checker

```python
def validate_transaction(amount):
    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    return amount, status


# Example
amount, status = validate_transaction(60000)
print("Transaction Amount:", amount)
print("Transaction Status:", status)
```

#Problem 9: Student Attendance Eligibility

```python
def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    for day in attendance_list:
        if day == 1:
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return percentage, status


# Example
attendance = [1,1,1,0,1,1,1,1,0,1]
percentage, status = check_exam_eligibility(attendance)
print("Attendance Percentage:", percentage)
print("Exam Eligibility:", status)
```

#Problem 10: Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):
    bill = 0

    # Slab Calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


# Example
bill, usage = calculate_electricity_bill(250)
print("Total Bill:", bill)
print("Usage Level:", usage)
```

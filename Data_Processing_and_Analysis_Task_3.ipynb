# Problem 1: Employee Performance Bonus Eligibility

```python

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Identify all employees with highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

# Problem 2: Search Query Keyword Analysis

```python
print("\n---- Problem 2: Search Query Keyword Analysis ----")

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation (if any)
import string
query = query.translate(str.maketrans('', '', string.punctuation))

words = query.split()

frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Display keywords searched more than once
result = {word: count for word, count in frequency.items() if count > 1}

print(result)
```

# Problem 3: Sensor Data Validation

```python
print("\n---- Problem 3: Sensor Data Validation ----")

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# Problem 4: Email Domain Usage Analysis

```python
print("\n---- Problem 4: Email Domain Usage Analysis ----")

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

# Problem 5: Sales Spike Detection


```python
print("\n---- Problem 5: Sales Spike Detection ----")

sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)

threshold = average_sales * 1.3  # 30% above average

for day, sale in enumerate(sales, start=1):
    if sale > threshold:
        print(f"Day {day}: {sale}")
```

# Problem 6: Duplicate User ID Detection

```python
print("\n---- Problem 6: Duplicate User ID Detection ----")

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

for uid in user_ids:
    id_count[uid] = id_count.get(uid, 0) + 1

for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
